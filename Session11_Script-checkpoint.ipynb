## RMD117 Images as Numpy Arrays

# Images as Numpy Arrays

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

Now we are ready to create a Numpy array and put some numbers in it. The Numpy array for storing an image is a height by width by three array of 8 bit, unsigned integers. That means they can store whole numbers from 0 to 255. Let's create a simple 200 by 300 rectangle with all the colours set to black.

```python
img1 = np.zeros((200,300,3),np.uint8)
plt.imshow(img1)
plt.title('All Black')
plt.show()
```

See how we defined the shape of the Numpy array to be 200 by 300 by three and set the data type to be 8 bit unsigned ints.

It is a bit minimalist as images go, but you can see how the shape of the Numpy array - 200 by 300 corresponds to the shape of the image - 200 pixels high and 300 wide. All pixels have a colour code of zero, zero, zero, which means no red, green or blue, so it is black. What happens if we change all the colour values for one colour only to 255? We should see a rectangle of pure red, green or blue. Let's try. We can do that with some Numpy slicing.

```python
img2 = np.zeros((200,300,3),np.uint8)
img2[:,:,2]=100

plt.imshow(img2)
plt.title('Image')
plt.show()
```

Here we set the first colour value to 255 for all pixels, leaving the others at zero. That gave us a red rectangle. Now I can change the zero in the slice to 1 to set the second colour value to 255. Now I get a green shape. Finally, if I change the third index in the slice to 2, then I set the third, blue value and get a blue shape.

So we now know that the order in which the matplotlib plotting code interprets the three colour values in a numpy array are red, green, and blue. If you are coding along in your own notebook. Pause the video here and try some other colour combinations.

Okay, now we can use some programming logic to set different pixels to different colours. Let's make a rectangle filled with a colour that graduates from red on the left to blue on the right. We do that by setting the green values all to zero and then setting the red value to start at 255 and descend to 0 while the blue value starts at zero and grows to 255 in every row from left to right. I'll run the code so you can see the output, and then we will take a look at what we did.

```python
img = [[[x,x,x] for x in range(256)] for i in range(256)]
img=np.array(img,np.uint8)
img[:,:,0]=255-img[:,:,0]
img[:,:,1]=0

plt.imshow(img)
plt.title('Image')
plt.show()
```

There are a few examples of Numpy usage here. The first line uses list comprehension to generate a 3D array - so that is an array of size 256, this outer one here, which contains arrays each of size 256 - the inner one here - which each contain an array of three values - the three exes here. The outer loop, indexed by i, generates a row of the image at a time. The next loop in, indexed by x, puts the values from 0 to 255 in each row of the image. Finally, the x,x,x array has the same value for each of the three colours.

Now, to grade from red to blue, we need to reverse the order of the red values so they go from 255 down to zero. The first of the two slices do that here. We take the first of the three colour values in every row, and every column and subtract its value from 255. Finally, the second slice sets the green values to zero. Can you see how that works? Take a moment to work it out if it is not immediately clear.

Hopefully that has given you an idea of how images are represented using Numpy. Perhaps the simplest image manipulation is scaling. We can make an image larger or smaller, or stretch one dimension using the resize function in OpenCV.

```python
height,width,_ = img.shape
bigger = cv2.resize(img,(2*width, 2*height))
print(bigger.shape)
```

We can rotate an image too, but that is a little more involved, as you can see here

```python
height,width,_ = bannf.shape
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
rotated_image = cv2.warpAffine(bannf, rotation_matrix, (width, height))
plt.imshow(rotated_image)
plt.show()
```

## RDB119 Drawing

# Drawing on an Image

Let's create a blank canvas to draw an image on. I'll create a Numpy array of 500 by 500 by 3, all full of ones and then use broadcasting to multiply every element in the array by 255. Now all the three-value colour codes contain 255, 255, 255, which is the code for white. I'll plot the image so you can see it.

```python
canvas = np.ones((500,500,3),np.uint8)*255
plt.imshow(canvas)
plt.show()
```

Well, there is not much to see, but you get the idea. Let's draw a red circle in the middle. Here it is. We want a circle, at location 250,250, of radius 200. The colour we want is defined by these three numbers: all red and no green or blue, and I have chosen a line thickness of 5.

```python
cv2.circle(canvas,(250,250),200,(255,0,0),5)   # Make negative thickness to fill
plt.imshow(canvas)
plt.show()
```

That is okay, but there are a couple of problems.The line is not very smooth and by default, matplotlib does not show the image full size. We can fix the second problem by embedding the image in a figure when we draw it. I'll write a function to do that so that we can use it repeatedly. Here it is, presented without comment.

```python
def display_image_full_size(img):
    height, width, _ = img.shape
    figsize = width/72, height/72
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.imshow(img)
    plt.show()
```

The problem of the line not being smooth can be solved with something called antialiasing. This smooths the edges of shapes by fading the colour from the line colour to the background colour at the edges of the shape. We specify that we want to use it by adding a line type argument like this. First, I make the canvas blank again, then I draw my circle with antialiasing.

```python
canvas = np.ones((500,500,3),np.uint8)*255
canvas = cv2.circle(canvas,(250,250),200,(250,20,20),5, lineType=cv2.LINE_AA)
display_image_full_size(canvas)
```

That is better. Now I will add a few more shapes in a similar way.

```python
canvas = cv2.line(canvas,(0,0),(500,500),(25,90,125),5, lineType=cv2.LINE_AA)
canvas = cv2.rectangle(canvas,(100,100),(200,200),(10,30,120),-1, lineType=cv2.LINE_AA)
canvas = cv2.ellipse(canvas,(256,256),(100,50),0,0,360,(20,200,20),-1, lineType=cv2.LINE_AA)
pts = np.array([[400,20],[450,70],[400,120],[350,70]], np.int32)
pts = pts.reshape((-1,1,2))
canvas = cv2.polylines(canvas,[pts],True,(80,255,200),3, lineType=cv2.LINE_AA)
display_image_full_size(canvas)
```

```python
display_image_full_size(canvas)
```

You can see that if I wanted to change the scale of the image, I would just make the numbers larger. This is useful if you have an image that does not come from a photograph that you want to use at different scales, always with perfect quality. A logo is a good example. ## RMD120 Edge Detection

# Edge Detection

Let's load an image with a single object in it and try to find the edges of that object.

```python
img = cv2.imread('Images/parrot.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Parrot')
plt.show()
```

Now we can clearly see what that is, but to the computer it is just an array of numbers. There is an edge detection algorithm, known as the Canny edge detector, which is named after John Canny, who invented it. It removes noise from an image and then calculates the change in brightness in both horizontal and vertical directions. It then chooses the largest differences and keeps those that are above a certain threshold. We run the Canny edge detector in Python with this code

```python
threshold1 = 100
threshold2 = 200
canny = cv2.Canny(img, threshold1, threshold2)

plt.imshow(canny, cmap='gray')
plt.show()
```

The thresholds determine how large a change in brightness needs to be to register as an edge. Jumps larger than the high threshold are all kept and jumps smaller than the low threshold are all removed. Those in between are only kept if they are attached to other lines that have been kept. You can see that it has done quite well, but I think we can improve it because there are still quite a few textured areas. We can try to remove those by blurring the image first. A so called median blurr passes a small window over the image and replaces the middle pixel with the median value of all the pixels in the window. This has the effect of blurring the image, but maintaining the edges. Here is the result of applying it to the parrot image.

```python
median = cv2.medianBlur(img,15)
plt.imshow(median)
plt.show()
```

In the example here, we use a window size of 15. You can experiment with different sizes to see what effect that has. See how the details are blurred, but the edges are still sharp. We can apply the edge detector to that.

```python
canny = cv2.Canny(median, threshold1, threshold2)
plt.imshow(canny, cmap='gray')
plt.show()
```

That is not bad, but there are a lot of gaps in the edge. We can fix that by dilating the image, which will join up the dots if they are close enough together. Let's do that now. 

```python
kernel = np.ones((3,3), np.uint8)  
dilate = cv2.dilate(canny, kernel, iterations=1)
plt.imshow(dilate)
plt.show()
```

That looks quite good doesn't it? You will have a chance to perform some adge detection of your own and share the results with your class mates. You can experiment with the thresholds and with different levels of blurring first.

## RMD122 Hough transforms for finding lines

# Hough Transforms for Finding Lines

Once you have found the edges, you can start to identify the shapes that the edges make. The simplest thing to look for is a straight line. 

There is a method that finds the straight lines in an image and returns their coordinates. The method uses something called Hough transforms. Let's use them to find the train tracks in a picture.

The python OpenCV function called HoughLinesP finds the lines in an image and returns their coordinates. We perform edge detection first, and then apply the Hough transform to find which of the edges represents a straight line. Here is the code.



```python
img = cv2.imread('Images/tracks.jpg', cv2.IMREAD_COLOR)
median = cv2.medianBlur(img,11)
gray = cv2.cvtColor(median, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 200)
lines = cv2.HoughLinesP(edges, 1, 0.001, 100, minLineLength=500, maxLineGap=100)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (255, 0, 0), 15)
plt.imshow(img)
plt.show()
```

We load the image, set its colour space to grayscale, find the edges with a Canny detector and then extract the lines by calling HoughLinesP.

See how the lines are returned as an array of start and end points for each line.

##  RDB 124 Haar Cascades

# Haar Cascades for Face Detection

We have seen how Haar cascades can be used to locate objects such as faces in an image. Now we will put that into practise using OpenCV in Python. We need to load an image, locate the faces, and then draw some boxes on the image to show where the faces were found. You have already learned how to do most of those things, but let's go through the code 

```python
img = cv2.imread('Images/prize.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

classifier = cv2.CascadeClassifier("haar/haarcascade_frontalface_default.xml")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = classifier.detectMultiScale(gray, 1.1, 5)
print(faces)
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
plt.imshow(img)
plt.show()
```

The classifier works on grayscale images, so we convert it and pass it to a function called detectMultiScale here. This function returns a list of coordinates and sizes of the boxes containing the images. We use those to draw boxes on the image to show where the faces were located. ## RMD 127 Finding Faces in a Webcam Feed

# Finding Faces in a Webcam Feed

Now we can process the image and show the processed version. Let's use the face detector we have already learned about to track my face in the image.

```python
cap = cv2.VideoCapture(0)
classifier = cv2.CascadeClassifier("haar/haarcascade_frontalface_default.xml")
cv2.namedWindow("frame", cv2.WINDOW_NORMAL)
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = classifier.detectMultiScale(gray, 1.1, 5)

    for (x,y,w,h) in faces:
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)

    # Display the resulting frame
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

## RMD 128 Magic Mirror

# Magic Mirror

```python
import cv2
import numpy as np


'''Move mask does the following
Any pixels that are different from the last frame are shown as the new frame
Any pixels that are the same as the previous frame fade towards the background
Background fade speed is argument sent to run mirror
'''


cap = cv2.VideoCapture(0)

def get_background():
    '''Capture background image, allowing the camera to settle'''
    for i in range(50):
        ret,background = cap.read()
    background = np.flip(background,axis=1)
    return background

def erode_dilate(mask):
    '''Remove isolated points of mask or holes in mask'''
    # Opening filter is used to remove small noisy dots of the transparency colour mask
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3,3),np.uint8),iterations=2)
    # Dilation fills in the gaps in larger areas of the mask
    mask = cv2.dilate(mask,np.ones((3,3),np.uint8),iterations = 1)
    return mask

def apply_mask(background, img, mask):
    '''Create an image where the masked part gets pixels from backgrounnd and the rest gets img
    mask is defined as image where each pixel is 0 (black) or not zero (e.g. 255 (white)).
    mask = 0 means output zero, mask !=0 means show image'''
    invmask = cv2.bitwise_not(mask)
	# Create an image that is the background in mask1 
    bg_part = cv2.bitwise_and(background,background,mask=mask)
    # Create an image that is the current background everywhere except in mask
    img_part = cv2.bitwise_and(img,img,mask=invmask)
    
    #Add the two images together
    final_output=bg_part+img_part
    return final_output

def move_mask(img):
    '''Compare the given image (which should be the previous frame) with a newly captured frame
    and create a mask of the differences
    
    return img, mask - img is the new frame directly from the camera and mask is white where the two frames differ and black
    everywhere else
    '''
    ret, img1 = cap.read()
    img1 = np.flip(img1,axis=1)
    dif=cv2.absdiff(img,img1)
    gray_image = cv2.cvtColor(dif, cv2.COLOR_BGR2GRAY)
    threshold = 10    # Threshold difference before a change is registered. Low means sensitive.
    ret,mask=cv2.threshold(gray_image,threshold,255,cv2.THRESH_BINARY_INV)
    # Erode and dilate the mask to fill small holes
    mask=erode_dilate(mask)
    return img1,mask


def run_mirror(bg=0.1):
    silent=0
    moved = 1-bg
    background=get_background()
    show=background
    ret, img = cap.read()
    img = np.flip(img,axis=1)
    while(cap.isOpened()):
        # Fetch one new frame (img) and a mask of pixels that have changed since the last one (mask)
        img,mask=move_mask(img)
        # Fade everything in the current image (show) towards the background
        show=cv2.addWeighted(background,bg,show,moved,0)
        
        # If there are very few changes over 100 iterations, assume the image is empty and take a new background
        minv = cv2.bitwise_not(mask)/255
        if np.sum(minv) < 5:
            silent+=1
        else:
            silent=0
        if silent == 100:
            ret,background = cap.read()
            background = np.flip(background,axis=1)
            silent=0
        
        # Use the mask of pixels that have changed to replace those in faded image with those from the latest frame
        show=apply_mask(show,img,mask)
    
        sized=cv2.resize(show,(1600,900))
        cv2.imshow('Magic Mirror',sized)
        k = cv2.waitKey(1)
        if k == 27:
            break

run_mirror(0.1)   
cap.release()
cv2.destroyAllWindows()
```

Here we are. If I move about, then I stay in the picture, but as soon as I stop, I start to vanish. If I just move my lips as I talk, or move my eyes like this, then that is all you can see. Let's stop that running and take a look at the code. I have put a lot of comments in it, so you can spend some time working it out for yourself, but here are a few points to note. The main function is called run mirror, and it takes one argument, which determines the speed of the fade of pixels that have not changed. The number determines the proportion of the background that is mixed into stationary pixels at each frame.
