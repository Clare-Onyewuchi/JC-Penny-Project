```python
#pandas is imported to help read the csv and json files
#index_col = 0 and index_col = 1 is indicated when reading the files 
#This tells pandas to treat the columns with index 0 and 1 as id where specified

import pandas as pd
users=pd.read_csv("C:/Users/clare/Downloads/users.csv", index_col=0)
display(users.head())

reviews=pd.read_csv("C:/Users/clare/Downloads/reviews.csv", index_col=1)
display(reviews.head())

products=pd.read_csv("C:/Users/clare/Downloads/products.csv", index_col=0)
display(products.head())

#lines=True parameter was used to read each line as a separate JSON object
#This is done because the JSON file is not in the standard arrray format
#orient='records' was used to handle JSON objects stored in a-line-delimeted format.

reviewers_json = pd.read_json("C:/Users/clare/Downloads/jcpenney_reviewers.json", orient='records',lines = True)
print(reviewers_json)
products_json = pd.read_json("C:/Users/clare/Downloads/jcpenney_products.json", orient='records', lines=True)
print(products_json.head(10))

# This is used to show the complete content of each column
pd.set_option('display.max_colwidth', None) 
```

# DATA MERGING

```python
#first merge
#the user and reviews csv file is merged using the merge function in pandas
#this merged files were merged and saved with a variable name - mergeUser_reviews
# how = 'left' is used to specify a left join
#on = 'Username' represents that these two files were merged using a common ID 'Username'
mergeUser_reviews = pd.merge(users, reviews, on='Username', how = 'left' )
display(mergeUser_reviews)

#second merge
#the merged data above was merged with a JSON file named reviewers_json
#they were merged using common ID's 'Username', 'DOB','State'outer join was applied - left
merged_data = pd.merge(reviewers_json, mergeUser_reviews, on = ['Username', 'DOB','State'],  how='left')
display(merged_data)


#the drop function was used to remove sku column from products_json 
#this was to avoid duplicates after being merged below
products_json_updated = products_json.drop(['sku', 'Reviews'], axis=1)
display(products_json_updated.head(10))


#third merge
#the products csv and products json file was merged using the merge function in pandas
#this merged files were merged and saved with a variable name called merged_data3
# how = 'outer' is used to specify an inner join
#on = 'Uniq_id', 'Name' represents that these two files were merged using common ID's
#head(10) was display only 10 observations in the table
#uniq_id in products-json and name_title was renamed to Uniq-id and Name respectively
merged_data3 = pd.merge(products, products_json_updated, on = ['Uniq_id', 'Name', 'average_product_rating'], how='left')
display(merged_data3.head(10))

#final merge
#the second and third merged data were merged together
#they were merged using a common ID named Uniq_id and a left join was applied
#head(10) was display only 10 observations in the table
final_merged_data1 = pd.merge(merged_data3, merged_data, on = 'Uniq_id', how='left')
final_merged_data = final_merged_data1.drop('Description', axis=1)
#the display was used to show the final merged data first 10 observations
display(final_merged_data.head(10))



```

```python
#imports the requests library to allow download of images from URLs
# imports BytesIO to enable the conversion of binary image data into a format suitable for the Image class 
# imports PIL to perform various image-related operations

import requests
from PIL import Image
from io import BytesIO

def analyze_image_size(image_url):
    try:
        # the image is downlaoded using the requests.get function
        response = requests.get(image_url)
        image_data = BytesIO(response.content)

        # PIL is used to open the image
        img = Image.open(image_data)

        # the size of the image is gotten and returned
        width, height = img.size
        return f"Image Size - Width: {width}, Height: {height}"
    
    except Exception as e:
        return f"Error analyzing image: {e}"

# for loop is used to Iterate over the dataset but limited it to just 50 images 
# it applies the analyze image clarity function to each image url in the product image urls column
# The results are then printed for each image.
for index, row in final_merged_data.head(50).iterrows():
    image_url = row['product_image_urls']
    result = analyze_image_size(image_url)
    print(f"For image at index {index}: {result}")

```

```python
# imports the OpenCV library, which is a popular computer vision and image
# imports the NumPy library. NumPy is a powerful library for numerical operations
#imports the requests library, which is a popular HTTP library in Python
#imports the Image class from the Pillow Library (PIL). 
# this used for opening, manipulating, and saving many different image file formats
# BytesIO is used to treat binary image data obtained from the HTTP response
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO

# the analyze image clarity function is used to take in an argument - image url
def analyze_image_clarity(image_url):
    try:
        # the requests library is used to download the image from the provided url
        response = requests.get(image_url)
        
        # BytesIO and Image.open from the PIL library is used to convert the downloaded image to a NumPy array
        image_data = BytesIO(response.content)
        img = Image.open(image_data)

        # the image is converted to a NumPy array
        img_np = np.array(img)

        # the image  is converted to grayscale using OpenCV's cv2.cvtColor function. 
        gray = cv2.cvtColor(img_np, cv2.COLOR_BGR2GRAY)

        # this calculates the laplacian grayscale image and the variance is computed using the var function
        laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
        
        # the result is returned as a string containing the calculated Laplacian variance.
        return f"Image Clarity and Sharpness - Laplacian Variance: {laplacian_var}"
    
    # an exception is added incase there is an error 
    except Exception as e:
        return f"Error analyzing image clarity: {e}"

# for loop is used to Iterate over the dataset but limited it to just 20 images 
#it applies the analyze image clarity function to each image in the image urls column
# The results are then printed for each image.
# a higher Laplacian Variance indicates more pronounced edges and in turn, higher sharpness.
for index, row in final_merged_data.head(20).iterrows():
    image_url = row['product_image_urls']
    result = analyze_image_clarity(image_url)
    print(f"For image at index {index}: {result}")
```

```python
import cv2
import numpy as np
import requests 
from PIL import Image
from io import BytesIO


def analyze_image_color(image_url):
    try:
        
        # the requests library is used to download the image from the provided URL.
        response = requests.get(image_url)
        image_data = BytesIO(response.content)

        # image.open is used to open the image
        img = Image.open(image_data)

        # the image is then converted to a NumPy array
        img_np = np.array(img)

        # the average colour of the image is computed 
        avg_color = img_np.mean(axis=(0, 1))
        
        
        # the average colour is converted to HSV color space
        avg_color_hsv = cv2.cvtColor(np.uint8([[avg_color]]), cv2.COLOR_BGR2HSV)[0][0]

        #  the image is checked to see if it has vibrant colors based on saturation 
        #and a threshold of 50 is set
        is_vibrant = avg_color_hsv[1] > 50 
        
        return f"Color Analysis - Average Color: {avg_color}"
    
    # an exception is added incase there is an error 
    except Exception as e:
        return f"Error analyzing image color: {e}"

# for loop is used to Iterate over the dataset but limited it to just 50 images 
# it applies the analyze image clarity function to each product image urls column
# The results are then printed for each image.
for index, row in final_merged_data.head(60).iterrows():
    image_url = row['product_image_urls']
    result = analyze_image_color(image_url)
    print(f"For image at index {index}: {result}")
    
```
